```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
generator(
    "Artificial intelligence is",
    max_length=30
)
```

```python
generator(
    "Once upon a time",
    max_length=40
)
```

```python
generator(
    "Today I learned",
    max_length=35
)
```

```python
generator(
    "The future of NLP",
    max_length=40
)
```

The model predicts one word at a time.

Different prompts produce different outputs.

Longer prompts usually produce more relevant text.

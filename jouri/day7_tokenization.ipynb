```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
sentences = [
    "I love NLP.",
    "ChatGPT helps students learn faster.",
    "Artificial Intelligence is changing the world."
]

for sentence in sentences:
    print("Sentence:", sentence)
    print(tokenizer.tokenize(sentence))
    print("-" * 50)
```

```python
short = "Hello"

long = "Hello, today I am learning Natural Language Processing using Hugging Face."

print("Short sentence tokens:")
print(tokenizer.tokenize(short))

print()

print("Long sentence tokens:")
print(tokenizer.tokenize(long))
```

```python
word = "internationalization"

print(tokenizer.tokenize(word))
```

```python
word = "unbelievable"

print(tokenizer.tokenize(word))
```

Today I learned that the tokenizer splits text into tokens before sending it to the model.

Long words may be divided into subwords.

Longer sentences usually produce more tokens.

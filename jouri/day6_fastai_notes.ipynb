## What is a Model?

A model is an AI program trained on data to make predictions.

## Dataset

A dataset is a collection of examples used to train a model.

## Training Set

The data used to teach the model.

## Validation Set

Data used to test the model after training.


## Accuracy

Accuracy is the percentage of correct predictions.

## Loss

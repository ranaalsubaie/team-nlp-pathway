```python
import gensim.downloader as api
```

```python
model = api.load("glove-wiki-gigaword-50")
```

```python
words = [
    "computer",
    "doctor",
    "school",
    "coffee",
    "book"
]

for word in words:
    print(f"\n{word}")
    print(model.most_similar(word, topn=5))
```

```python
model.most_similar(
    positive=["woman","king"],
    negative=["man"],
    topn=5
)
```

```python
model.most_similar(
    positive=["Paris","Italy"],
    negative=["France"],
    topn=5
)
```

```python
model.most_similar(
    positive=["walking","swimming"],
    negative=["walk"],
    topn=5
)
```

Word embeddings represent words as vectors. Similar words have similar vectors. GloVe helps us find semantic relationships between words.
